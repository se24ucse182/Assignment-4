```python
districts = [
'Adilabad','Bhadradri','Hyderabad','Jagtial','Jangaon','Jayashankar',
'Jogulamba','Kamareddy','Karimnagar','Khammam','Komaram Bheem',
'Mahabubabad','Mahabubnagar','Mancherial','Medak','Medchal',
'Mulugu','Nagarkurnool','Nalgonda','Narayanpet','Nirmal','Nizamabad',
'Peddapalli','Rajanna','Rangareddy','Sangareddy','Siddipet',
'Suryapet','Vikarabad','Wanaparthy','Warangal Rural','Warangal Urban','Yadadri'
]

neighbors = {
'Adilabad':['Komaram Bheem','Nirmal'],
'Komaram Bheem':['Adilabad','Mancherial'],
'Mancherial':['Komaram Bheem','Peddapalli'],
'Peddapalli':['Mancherial','Karimnagar'],
'Karimnagar':['Peddapalli','Rajanna'],
'Rajanna':['Karimnagar','Jagtial'],
'Jagtial':['Rajanna','Nizamabad'],
'Nizamabad':['Jagtial','Kamareddy'],
'Kamareddy':['Nizamabad','Medak'],
'Medak':['Kamareddy','Sangareddy'],
'Sangareddy':['Medak','Rangareddy'],
'Rangareddy':['Sangareddy','Hyderabad','Medchal'],
'Hyderabad':['Rangareddy','Medchal'],
'Medchal':['Hyderabad','Rangareddy','Siddipet'],
'Siddipet':['Medchal','Jangaon'],
'Jangaon':['Siddipet','Warangal Urban'],
'Warangal Urban':['Jangaon','Warangal Rural'],
'Warangal Rural':['Warangal Urban','Jayashankar'],
'Jayashankar':['Warangal Rural','Mulugu'],
'Mulugu':['Jayashankar','Bhadradri'],
'Bhadradri':['Mulugu','Khammam'],
'Khammam':['Bhadradri','Suryapet'],
'Suryapet':['Khammam','Nalgonda'],
'Nalgonda':['Suryapet','Yadadri'],
'Yadadri':['Nalgonda','Medchal'],
'Mahabubabad':['Warangal Rural','Khammam'],
'Mahabubnagar':['Nagarkurnool','Wanaparthy'],
'Nagarkurnool':['Mahabubnagar','Wanaparthy'],
'Wanaparthy':['Nagarkurnool','Mahabubnagar'],
'Narayanpet':['Mahabubnagar'],
'Vikarabad':['Rangareddy'],
'Nirmal':['Adilabad','Nizamabad'],
'Jogulamba':['Mahabubnagar']
}

colors = ['Red','Green','Blue','Yellow']

def is_valid(district, color, assignment):
    for n in neighbors.get(district, []):
        if n in assignment and assignment[n] == color:
            return False
    return True

def backtrack(assignment):
    if len(assignment) == len(districts):
        return assignment
    unassigned = [d for d in districts if d not in assignment][0]
    for color in colors:
        if is_valid(unassigned, color, assignment):
            assignment[unassigned] = color
            result = backtrack(assignment)
            if result:
                return result
            del assignment[unassigned]
    return None

solution = backtrack({})

# Simple clean output
for d in districts:
    print(f"{d} -> {solution[d]}")
```
