```python
import itertools

def solve_cryptarithm(puzzle):
    words = puzzle.replace(' ', '').split('=')
    left = words[0].split('+')
    right = words[1]

    letters = set(''.join(left) + right)
    letters = list(letters)

    if len(letters) > 10:
        print("Too many unique letters")
        return

    first_letters = set(word[0] for word in left + [right])

    digits = range(10)

    for perm in itertools.permutations(digits, len(letters)):
        mapping = dict(zip(letters, perm))

        if any(mapping[ch] == 0 for ch in first_letters):
            continue

        left_values = []
        for word in left:
            num = int(''.join(str(mapping[ch]) for ch in word))
            left_values.append(num)

        right_value = int(''.join(str(mapping[ch]) for ch in right))

        if sum(left_values) == right_value:
            print("Solution Found:")
            print(mapping)
            for word in left:
                print(''.join(str(mapping[ch]) for ch in word), end=' + ')
            print("\b\b= ", right_value)
            return

    print("No solution found")


puzzle = "SEND + MORE = MONEY"
solve_cryptarithm(puzzle)
```
